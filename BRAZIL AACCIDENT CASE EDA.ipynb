Calculaing Z-scores
Z score value is above the value which we get after calculating (data-data(mean))/satandared deviation of data
we want to a threshold value while calculating z_score usually as a stndarad the threshold value is taken as 3

```python
df['z_score']=(df['latitude']- df['latitude'].mean())/df['latitude'].std()
threshold=2
df_no_outliers=df[df['z_score'].abs()<=threshold]
df_no_outliers=df_no_outliers.drop(columns=['z_score'])
print(df_no_outliers.head())
```
